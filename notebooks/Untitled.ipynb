```python
import gym
from torchvision import transforms
from utils.misc import LSIZE, RED_SIZE
from models.vae import VAE
from utils.explore_misc import RolloutGeneratorSingle
```

```python
vae_dir = '/media/rene/data/carracingWM/white_10000/vae'
device = "cuda:1"
model = VAE(3, LSIZE).to(device)

reload_file = os.path.join(vae_dir, 'best.tar')
state = torch.load(reload_file)
model.load_state_dict(state['state_dict'])            


transform_test = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((RED_SIZE, RED_SIZE)),
    transforms.ToTensor(),
])
```

```python
loc = '/media/rene/data/carracingWM/brown_10000/train/thread_0/rollout_1.npz'
img = np.load(loc)['observations'][100, :, :, :]

obs = transform_test(img).unsqueeze(0).to(device)
out = model(obs)


# img = np.moveaxis(img, 0, 2)
plt.figure(figsize=(8,8))
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.imshow(img)


out = np.squeeze(out[0].data.cpu().numpy())
out = np.moveaxis(out, 0, 2)
plt.figure(figsize=(8,8))
plt.title('Reconstruction')
plt.xticks([])
plt.yticks([])
plt.imshow(out)
```
